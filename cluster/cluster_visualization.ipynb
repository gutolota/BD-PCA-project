```python
import pandas as pd
from sklearn.cluster import KMeans
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv', sep=',')
```

```python
k_model = KMeans(n_clusters=2, max_iter=10000000000)
k_model.fit(df)
data = k_model.predict(df)
data_df = df.copy()
data_df['cluster'] = data
```

```python
data_df
```

```python
data_df.info()
```

```python
plt.figure(figsize=(11, 8))
sns.scatterplot(data=data_df, x='0', y='1', hue='cluster')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['A', 'B'], title='cluster')
```

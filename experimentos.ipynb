```python
import random
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

### Exemplo de cluster

```python
centros = [[2, 5], [6, 2]]
data = np.array([[random.gauss(centros[random.randint(0, 1)][0], 0.4), random.gauss(centros[random.randint(0, 1)][1], 0.4)] for _ in range(1000)])
```

```python
clusters = KMeans(n_clusters=4).fit_predict(data)
```

```python
plt.figure(figsize=(12, 9))
sns.set_theme()
sns.scatterplot(x=data[:, 0], y=data[:, 1], hue=clusters)
plt.legend(['A', 'B', 'C', 'D'], title='Cluster')
```
